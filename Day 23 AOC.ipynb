# Bruteforce

```python
import copy
```

```python
inp = inp.replace("D", "4")
inp = inp.replace("C", "3")
inp = inp.replace("B", "2")
inp = inp.replace("A", "1")
inp = inp.replace("#", " ")
split = [n.strip().split(" ") for n in inp.split("\n")[2:4]]
print(split)
pos = [[],[],[],[],[],[],[],[],[],[],[]]
for i in [1,0]:
    for j in range(4): pos[2+2*j] += [int(split[i][j])]
print(pos)
```

```python
def cost(q,from_here,to_here):
    mult = 10**(q[from_here][-1]-1)
    dist = abs(from_here-to_here)
    if from_here in [2,4,6,8]: dist += 3-len(q[from_here])
    if to_here in [2,4,6,8]: dist += 2-len(q[to_here])
    return dist*mult
```

```python
def reachable(q,i):
    r = []
    k = i-1
    while True:
        r += [k]
        if k not in [2,4,6,8] and len(q[k]) == 1: break
        elif k == 0: break
        k -= 1
    k = i+1
    while True:
        r += [k]
        if k not in [2,4,6,8] and len(q[k]) == 1: break
        elif k == 10: break
        k += 1
    return r
```

```python
def solve(q, costs):
    #print("NEW")
    cont = True
    while cont:
        #print(q)
        cont = False
        done = []
        for i in [2,4,6,8]:
            # Check if any stack is finished and fill
            if all([j == i//2 for j in q[:][i]]):
                done += [i]
                r = reachable(q,i)
                for e in r:
                    if len(q[e])>0 and q[e][-1] == i//2:
                        costs += cost(q,e,i)
                        q[i].append(q[e].pop())
                        cont = True
    
    if len(done) == 4 and sum([len(q[n]) for n in range(11) if n not in [2,4,6,8]]) == 0: 
            return q, costs, True
        
    # Expand new paths  
    best_costs = 10**9
    for i in [2,4,6,8]:
        if i not in done:
            r = reachable(q,i)
            for e in r:
                if len(q[e]) == 0 and len(q[i]) > 0 and e not in [2,4,6,8]:
                    new_q = copy.deepcopy(q)
                    new_cost = costs + cost(q,i,e)
                    new_q[e].append(new_q[i].pop())
                    q_alt, costs_alt, finished = solve(new_q, new_cost)
                    if finished: 
                        #print(costs_alt)
                        if costs_alt < best_costs:
                            best_costs = costs_alt
                            best_q = copy.deepcopy(q_alt)
                            
    if best_costs < 10**9:
        return best_q, best_costs, True
    else:            
        return q, costs, False #If nothing worked
```
